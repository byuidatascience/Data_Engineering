```python
import boto
import pandas as pd


conn = boto.connect_s3()
mybucket = conn.get_bucket('tripaul-bucket')
f = mybucket.get_key('shakespeare-sonnets.txt')
data = str(f.get_contents_as_string())
print (data)
```

```python
import re

data = data.replace('\\n','')
data = data.lower()
data = re.sub('[^A-Za-z0-9]+', ' ', data)
```

```python
print (data)
```

```python
from collections import Counter
import pandas as pd

splitted = data.split()

c = Counter(splitted)
```

```python
import collections
```

```python
this = collections.OrderedDict(c.most_common())
```

```python
print (c.most_common(20)) # Most common 20.
```

```python
web = pd.DataFrame.from_dict(this, 'index')
```

```python
web.to_html('upload_me_now.html')
```
